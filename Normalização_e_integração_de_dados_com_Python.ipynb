```python
%pip install pandas requests numpy lxml beautifulsoup4 html5lib

import requests
import pandas as pd
import numpy as np
import io

#link de pesquisa
url = "https://www.espn.com.br/futebol/estatisticas/_/liga/bra.1"
headers = {"User-Agent":"Mozilla/5.0"}

resp_wiki = requests.get(url, headers=headers)
resp_wiki.raise_for_status()

#comando de leitura
tables_wiki = pd.read_html(io.StringIO(resp_wiki.text))

#tabelas encontradas no link
if len(tables_wiki) > 0:
    df_copa = tables_wiki[0]
    print(f"Tabelas encontradas: {len(tables_wiki)}")
else:
    print("Nenhuma tabela encontrada na página.")

#nomeação de tabelas
artilheiros = tables_wiki[0]
assistencias = tables_wiki[1]

#alteração dos nomes das colunas
artilheiros.rename(columns={
    'Nome': 'Jogadores',
    'Hora': 'Clube',
    'P': 'Partidas',
    'G': 'Gols'
}, inplace=True)

assistencias.rename(columns={
    'Nome': 'Jogadores',
    'Hora': 'Clube',
    'P': 'Partidas',
    'G': 'Assistências',
    'A': 'Assistências'
}, inplace=True)


#exclusão de coluna
artilheiros.drop('POS', axis=1, inplace=True)
assistencias.drop('POS', axis=1, inplace=True)


#impressão das tabelas
print("=== Artilheiros ===")
print(artilheiros.head())

print("\n=== Assistencias ===")
print(assistencias.head())

artilheiros.to_csv('artilheiros.csv', index=False)
assistencias.to_csv('assistencias.csv', index=False)

```
