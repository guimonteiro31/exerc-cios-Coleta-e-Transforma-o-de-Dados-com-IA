```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import io # Importar StringIO

url = 'https://pt.wikipedia.org/wiki/Lista_de_presidentes_do_Brasil'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status() # Lança um erro para status de requisição HTTP ruins (4xx ou 5xx)

    # Passar o conteúdo HTML bruto para pd.read_html. Isso é mais robusto para tabelas complexas.
    tables = pd.read_html(io.StringIO(response.text)) # Usar StringIO para conteúdo HTML literal

    df = None
    found_president_col = False # Flag para indicar se uma coluna de presidente foi encontrada

    # Iterar sobre os DataFrames para encontrar a tabela correta
    for i, table_df in enumerate(tables):
        print(f"--- Verificando tabela {i} ---")
        print("Colunas encontradas:", table_df.columns.tolist()) # Imprimir colunas para inspeção

        # Verificar a existência de 'Presidente' em várias formas
        for col_name in table_df.columns:
            # Lidar com colunas de multi-índice (tuplas)
            if isinstance(col_name, tuple):
                # Aplanar os componentes da tupla para correspondência de string
                col_name_str = ' '.join(filter(None, [str(c) for c in col_name])).lower()
            else:
                col_name_str = str(col_name).lower()

            if 'presidente' in col_name_str:
                df = table_df
                found_president_col = True
                print(f"Tabela de presidentes encontrada no índice {i} com coluna contendo: '{col_name}'.")
                display(df.head())
                break # Sair do loop interno assim que a coluna de presidente for encontrada

        if found_president_col:
            break # Sair do loop externo assim que a tabela correta for encontrada

    if df is None:
        print("Não foi possível encontrar a tabela de presidentes na página.")
    else:
        print("DataFrame com a lista de presidentes (poderá necessitar de ajuste de colunas):")
        display(df.head())


except requests.exceptions.RequestException as e:
    print(f"Ocorreu um erro na requisição HTTP: {e}")
except Exception as e:
    print(f"Ocorreu um erro geral ao tentar ler as tabelas: {e}")
```

Com a tabela extraída, o primeiro passo é limpar os nomes das colunas. As colunas estão em um formato de multi-índice, o que pode dificultar a manipulação. Vamos simplificá-las para que fiquem mais acessíveis.

```python
import re # Ensure re is imported for regex operations

# Re-initialize df with the correct table from the 'tables' list (identified in f5ccf19b as index 0)
df = tables[0]

# Limpar os nomes das colunas

print(f"DEBUG: Initial df columns: {df.columns.tolist()}\n")

# 1. Aplanar o MultiIndex de forma robusta
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)
    print(f"DEBUG: df columns after flattening MultiIndex: {df.columns.tolist()}\n")
else:
    print(f"DEBUG: df columns was not MultiIndex, current: {df.columns.tolist()}\n")

# 2. Limpar strings de nomes de colunas (remover conteúdo entre parênteses e espaços extras)
new_columns_step2 = []
for col_name in df.columns:
    cleaned_name = re.sub(r'\s*\(.*\)', '', str(col_name)).strip()
    new_columns_step2.append(cleaned_name)
df.columns = new_columns_step2
print(f"DEBUG: df columns after string cleaning (removing parenthesis content): {df.columns.tolist()}\n")

# 3. Renomear as colunas para termos nomes finais mais simples e consistentes
# O mapeamento considera os nomes exatos esperados após o passo 2 e os novos nomes desejados.
column_rename_map = {
    'N.°': 'Numero',
    'Presidente': 'Presidente',
    'Período do mandato': 'Periodo_Mandato',
    'Fotografia': 'Fotografia',
    'Partido': 'Partido',
    'Vice-presidente': 'Vice_Presidente',
    'Eleição': 'Eleicao',
    'Notas e referências': 'Notas_Referencias'
}

# Aplicar o renomeamento para as colunas que existem no DataFrame após o passo 2
df = df.rename(columns={old_name: new_name for old_name, new_name in column_rename_map.items() if old_name in df.columns})
print(f"DEBUG: df columns after final renaming step: {df.columns.tolist()}\n")

# 4. Filtrar e selecionar apenas as colunas que realmente precisamos
final_desired_columns = ['Numero', 'Presidente', 'Periodo_Mandato', 'Partido', 'Vice_Presidente', 'Eleicao']

# Filtrar o DataFrame para manter apenas as colunas desejadas que existam após o renomeamento
df = df.filter(items=final_desired_columns)
print(f"DEBUG: df columns after final filtering: {df.columns.tolist()}\n")

# Exibir as novas colunas e as primeiras linhas do DataFrame limpo
print("Colunas do DataFrame após limpeza e seleção final:")
print(df.columns.tolist())
print("DataFrame após limpeza das colunas:")
display(df.head())
```

Agora que as colunas estão limpas, precisamos extrair o ano de início do mandato da coluna `Periodo_Mandato`. Essa coluna contém informações de data e duração, e nosso objetivo é obter apenas o ano para calcular a década. Também vamos extrair o nome do presidente de forma mais limpa.

```python
# Extrair o ano de início do mandato
def extract_start_year(periodo_mandato):
    if pd.isna(periodo_mandato):
        return None
    # Tenta encontrar um padrão como 'dia de mês de ANO' no início da string
    import re
    match = re.search(r'\d{1,2}\s+de\s+\w+\s+de\s+(\d{4})', periodo_mandato)
    if match:
        return int(match.group(1))
    # Se não encontrar, tenta encontrar o primeiro número de 4 dígitos
    match = re.search(r'(\d{4})', periodo_mandato)
    if match:
        return int(match.group(1))
    return None

# Adicionar uma verificação de coluna para robustez
periodo_col_name = None
if 'Periodo_Mandato' in df.columns:
    periodo_col_name = 'Periodo_Mandato'
elif 'Período do mandato' in df.columns:
    periodo_col_name = 'Período do mandato'

if periodo_col_name:
    df['Ano_Inicio'] = df[periodo_col_name].apply(extract_start_year)
else:
    print("Erro: Nenhuma das colunas 'Periodo_Mandato' ou 'Período do mandato' foi encontrada.")
    df['Ano_Inicio'] = None # Definir como None ou NaN para evitar erros posteriores

# Limpar o nome do presidente, removendo o que está entre parênteses e espaços extras
def clean_president_name(name):
    if pd.isna(name):
        return None
    import re
    cleaned_name = re.sub(r'\s*\(.*\)', '', name).strip()
    # Algumas entradas repetem o nome com informações adicionais, vamos pegar apenas o primeiro nome
    return cleaned_name.split('\n')[0].strip()

df['Presidente_Limpo'] = df['Presidente'].apply(clean_president_name)

# Remover linhas completamente duplicadas (mantendo o primeiro mandato)
# Isso ajuda a evitar contagens duplas para presidentes que apareceram mais de uma vez na mesma linha na tabela original
df_cleaned = df.drop_duplicates(subset=['Presidente_Limpo', 'Ano_Inicio', periodo_col_name])

# Filtrar e reorganizar as colunas conforme solicitado pelo usuário
final_selected_columns = ['Presidente_Limpo', periodo_col_name, 'Ano_Inicio']
df_cleaned = df_cleaned[final_selected_columns]

# Exibir as colunas relevantes e as primeiras linhas do DataFrame com os anos e nomes limpos
print("DataFrame com Ano_Inicio e Presidente_Limpo:")
display(df_cleaned.head(10))
```

Agora que temos os anos de início dos mandatos, podemos criar a coluna da década e agrupar os presidentes para obter as tabelas solicitadas.

```python
# Criar a coluna 'Decada'
df_cleaned['Decada'] = (df_cleaned['Ano_Inicio'] // 10) * 10

# Tabela 1: Presidentes agrupados por década
# Vamos criar uma lista de presidentes para cada década
presidentes_por_decada = df_cleaned.groupby('Decada')['Presidente_Limpo'].apply(lambda x: list(set(x.dropna()))).reset_index()
print("### Presidentes agrupados por Década ###")
display(presidentes_por_decada)

# Tabela 2: Contagem de vezes que sofreram presidentes (contagem de mandatos)
# Assumindo que cada linha única de Presidente_Limpo e Ano_Inicio representa um mandato distinto
contagem_mandatos = df_cleaned.groupby('Presidente_Limpo').size().reset_index(name='Numero_Mandatos')
contagem_mandatos = contagem_mandatos.sort_values(by='Numero_Mandatos', ascending=False)
print("\n### Contagem de Mandatos por Presidente ###")
display(contagem_mandatos.head(10))
```
